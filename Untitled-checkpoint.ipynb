IMPORTING NECESSARY LIBRARIES

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

TRAIN DATA

```python
data_train=pd.read_csv('trainset.csv.txt')
data_train.head()
```

OPEN PRICE - VOLUME VS TIME PLOT

```python
fig,ax=plt.subplots(nrows=2,sharex=True,figsize=(15,8))
ax[0].plot(data_train['Date'],data_train['Open'])
ax[1].plot(data_train['Date'],data_train['Volume'])
ax[0].set_title('Open-Price')
ax[1].set_title('Volume')
```

DATA PREPROCESSING


```python
data_train.shape
```

```python
data_train['Date']=pd.to_datetime(data_train.Date,format='%Y-%m-%d')
data_train.index=data_train['Date']
```

```python
open_train=data_train.iloc[:,1:2].values
open_train
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
open_train_sc=scaler.fit_transform(open_train)
open_train_sc
```

```python
X_train=[]
y_train=[]

for i in range(50,1258):
    X_train.append(open_train_sc[i-50:i,0])
    y_train.append(open_train_sc[i,0])
```

```python
X_train
```

```python
X_train=np.array(X_train)
y_train=np.array(y_train)
X_train
```

```python
a,b,c=X_train.shape[0],X_train.shape[1],1
X_train=np.reshape(X_train,(a,b,c))
```

```python
X_train.shape
```

IMPORTING KERAS AND BUILDING THE MODEL

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
regressor=Sequential()
regressor.add(LSTM(units=20,input_shape=(None,1)))
regressor.add(Dropout(0.2))
regressor.add(Dense(units=1))
```

```python
regressor.compile(optimizer='adam',loss='mean_squared_error')
regressor.fit(X_train,y_train,epochs=100,batch_size=32)
```

TEST DATA

```python
data_test=pd.read_csv('testset.csv.txt')
real_price=data_test.iloc[:,1:2].values
```

```python
data_total=pd.concat((data_train['Open'],data_test['Open']),axis=0)
inputs=data_total[len(data_total)-len(data_test)-60:].values
inputs=inputs.reshape(-1,1)
inputs=scaler.transform(inputs)
```

```python
data_test['Date']=pd.to_datetime(data_test.Date,format='%Y-%m-%d')
data_test.index=data_test['Date']
data_test.shape
```

```python
X_test=[]
for i in range(60,185):
    X_test.append(inputs[i-60:i,0])
X_test=np.array(X_test)
a,b,c=X_test.shape[0],X_test.shape[1],1
X_test=np.reshape(X_test,(a,b,c))
```

PREDICTION

```python
predicted_price=regressor.predict(X_test)
predicted_price=scaler.inverse_transform(predicted_price)
```

```python
pred=pd.DataFrame(predicted_price)
pred.index=data_test['Date']
```

PREDICTED VS REAL PLOTS

```python
plt.figure(figsize=(15,8))
plt.plot(data_train['Open'],color='red',label='PAST GOOGLE STOCK PRICE')
plt.plot(data_test['Open'],color='blue',label='FUTURE GOOGLE STOCK PRICE ( ACTUAL)')
plt.plot(pred,color='green',label='FUTURE GOOGLE STOCK PRICE (PREDICTED)')
plt.title('GOOGLE STOCK PRICE PREDICTION')
plt.xlabel('Time')
plt.ylabel('Google Stock Price')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.plot(real_price,color='blue',label='GOOGLE STOCK PRICE (ACTUAL)')
plt.plot(predicted_price,color='red',label='GOOGLE STOCK PRICE (PREDICTED)')
plt.title('GOOGLE STOCK PRICE PREDICTION')
plt.xlabel('Time')
plt.ylabel('Google stock price')
plt.legend()
plt.show()
```
